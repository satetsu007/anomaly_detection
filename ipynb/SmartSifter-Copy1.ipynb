## SDLEアルゴリズム

```python
random.seed(0)
T = 100000
x = np.array([math.floor(random.random()*6) for i in range(T)])
```

```python
class SDLE:
    def __init__(self, r, beta, M, T):
        """
        """
        self.r = r
        self.beta = beta
        self.M = M
        self.T = T
        
        self.count = np.zeros((T+1, M))
        self.T_t = np.zeros((T+1, M))
        self.prob = np.zeros((T+1, M))
        
        self.t = 1
    
    def update(self, x_t):
        """
        オンライン学習
        """
        for i in range(self.M):
            delta = 0
            if i == x_t:
                delta = 1

            self.T_t[self.t, i] = (1 - self.r) * self.T_t[self.t-1, i] + delta
            self.count[self.t, i] = self.count[self.t-1, i] + delta
            if self.r == 0:
                q = (self.T_t[self.t, i] + self.beta) / (self.t + self.M * self.beta)
                # self.prob[self.t, i] = q / self.count[self.t, i]
                # self.prob[self.t, i] = q
            else:
                q = (self.T_t[self.t, i] + self.beta) / ((1 - (1 - self.r) ** self.t) / self.r + self.M * self.beta)
                # self.prob[self.t, i] = q
                # self.prob[self.t, i] = q / self.count[self.t, i]
            
            if not self.count[self.t, i] == 0:
                self.prob[self.t, i] = q / self.count[self.t, i]

        self.t += 1
    
    def train(self, x):
        """
        バッチ学習
        """
        while self.t <= self.T:
            self.update(x[self.t-1])
```

```python
r = 0.1
beta = 1
M = len(np.unique(x))

sdle = SDLE(r, beta, M, T)
# print("t = %s" % sdle.t)
# print("count:\n", sdle.count)
# print("prob:\n",sdle.prob)
```

```python
class SDEM:
    def __init__(self, r, alpha, K, T, d):
        """
        """
        self.r = r
        self.alpha = alpha
        self.K = K
        self.T = T + 1
        self.d = d
        
        self.mu = np.zeros((self.T, self.K, self.d))
        self.mu_ = np.zeros((self.T, self.K, self.d))
        self.c = np.zeros((self.T, self.K))
        self.Lambda = np.zeros((self.T, self.K, self.d, self.d))
        self.Lambda_ = np.zeros((self.T, self.K, self.d, self.d))
        
        min_x, max_x = y.T[0].min(), y.T[0].max()
        min_y, max_y = y.T[1].min(), y.T[1].max()
        
        for i in range(self.K):
            self.mu_[0] = np.c_[np.random.uniform(min_x, max_x, self.K), np.random.uniform(min_y, max_y, self.K)]
            self.mu[0] = np.copy(self.mu_[0])
            if i == self.K-1:
                self.c[0, i] = 1 - (i / self.K)
            else:
                self.c[0, i] = 1 / self.K
            self.Lambda_[0, i] = np.identity(self.d) * 0.1
            self.Lambda[0, i] = np.copy(self.Lambda_[0, i])
        
        self.t = 1
    
    def calc_likelihood(self, idy):
        likelihood = np.zeros(self.K)
        for k in range(K):
            likelihood[k] = st.multivariate_normal.pdf(idy, self.mu[self.t-1, k], self.Lambda[self.t-1, k])
        return likelihood
    
    def update(self, idy):
        """
        オンライン学習
        """
        for i in range(self.K):
            t = self.t
            p = self.calc_likelihood(idy)
            global gamma
            gamma = (1 - self.alpha * self.r) * (self.c[t-1, i] * p[i]) / (np.dot(self.c[t-1], p))\
                                                                            + (self.alpha * self.r) / self.K
            
            self.c[t, i] = ((1 - self.r) * self.c[t-1, i]) + (self.r * gamma)
            self.mu_[t, i] = ((1 - self.r) * self.mu_[t-1, i]) + (self.r * gamma * idy)
            self.mu[t, i] = self.mu_[t, i] / self.c[t, i]
            
            idy_tmp = idy.reshape((2, 1))
            mu_tmp = self.mu[t, i].reshape((2, 1))
            
            self.Lambda_[t, i] = ((1 - self.r) * self.Lambda_[t-1, i]) + (self.r * gamma * np.dot(idy_tmp, idy_tmp.T))
            self.Lambda[t, i] = (self.Lambda_[t, i] / self.c[t, i])# - (np.dot(mu_tmp, mu_tmp.T))
            
        self.t += 1
    
    def train(self, y):
        """
        バッチ学習
        """
        while self.t < self.T:
            self.update(y[self.t-1])
    
    def details(self, idy):
        i = self.t-1
        print("----------------t=%s----------------" % i)
        print("mu: \n", sdem.mu[i])
        print("mu_: \n", sdem.mu_[i])
        print("c: \n", sdem.c[i])
        print("Lambda: \n", sdem.Lambda[i])
        print("Lambda_: \n", sdem.Lambda_[i])
        print("sum of log likelihood: \n", np.sum(np.log(sdem.calc_likelihood(idy))))
```

```python
# set parameter
r = 1 / y.shape[0]
alpha = 1.0
K = 3
T = y.shape[0]
d = y.shape[1]
```
