```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import math
import seaborn as sns
```

```python
random.seed(0)
T = 1000000
x = np.array([math.floor(random.random()*6) for i in range(T)])
```

```python
plt.hist(x, rwidth=0.9)
plt.show()
```

```python
r = 0.001
beta = 1
M = 6
X = range(M)
T_t = np.zeros(M)
T_t_1 = np.zeros(M)
A = np.zeros(M)
t = 1
prob = np.zeros((T, M))
```

```python
while t <= T:
    for i in X:
        delta = 0
        if i == x[t-1]:
            delta = 1
        A[i] += delta
            
        T_t[i] = (1-r)*T_t_1[i]+delta
        q = (T_t[i]+beta)/((1-(1-r)**t)/r+M*beta)
    
        if not A[i] == 0:
            prob[t-1, i] = q/A[i]
    T_t_1 = np.copy(T_t)
    t += 1
```

```python
p = prob[-1] * (1/prob[-1].sum())
```

```python
plt.bar(range(6), p)
plt.show()
```
