```python
#-*- coding:utf-8 -*-
import numpy as np
import numpy.random as rd
import scipy as sp
from scipy import stats as st

import matplotlib.pyplot as plt
```

# 混合ガウス分布に従うデータの生成

```python
seed = 0
n = [200, 150, 150] #各データ数
K = 3 #潜在変数の数
D = 2 #次元

#mu:D次元
mu_true = np.array(
    [[0.2, 0.5],
     [1.2, 0.5],
     [2.0, 0.5]])

#sigma: D×D次元
sigma_true = np.array(
    [[[0.1,  0.085], [0.085, 0.1]],
     [[0.1, -0.085], [-0.085, 0.1]],
     [[0.1,  0.085], [0.085, 0.1]]
    ])

rd.seed(seed)
org_data = None
for i in range(K):
    #k_0 に属するデータを生成
    if org_data is None:
        org_data = np.c_[st.multivariate_normal.rvs(mean=mu_true[i], cov=sigma_true[i], size=n[i]), np.ones(n[i])*i]
        
    #k_1, k_2に属するデータを生成し、結合する
    else:
        tmp_data = np.c_[st.multivariate_normal.rvs(mean=mu_true[i], cov=sigma_true[i], size=n[i]), np.ones(n[i])*i]
        org_data = np.r_[org_data, tmp_data]

print(org_data)
```

```python
# グラフ作成
fig = plt.figure(figsize=(12, 5))

#グラフ描画
c = ['r', 'g', 'b']
for i in range(K):
    plt.scatter(org_data[org_data[:,2]==i][:,0], org_data[org_data[:,2]==i][:,1], c=c[i])

plt.show()
```

# EMアルゴリズム

## 関数の定義

```python
def calc_likelihood(data, mu, sigma, pi, K):
    '''対数尤度を計算する'''
    
    pi_Ns = np.zeros((len(data), K))
    for k in range(K):
        pi_Ns[:, k] = [pi[k] * st.multivariate_normal.pdf(d, mu[k], sigma[k]) for d in data]
    
    log_likelihood = np.sum(np.log(np.sum(pi_Ns, axis=1)))
    
    return pi_Ns, log_likelihood
```

## 1. パラメータの初期化

```python
def initialize(data, K, D):
    '''パラメータを初期化する関数'''
    
    #piの初期化
    pi = np.zeros(K)
    for k in range(K):
        if k == K-1:
            pi[k] = 1 - np.sum(pi)
        else:
            pi[k] = 1 / K
    
    #muの初期化
    max_x, min_x = np.max(data[:,0]), np.min(data[:,0])
    max_y, min_y = np.max(data[:,1]), np.min(data[:,1])
    mu = np.c_[rd.uniform(low=min_x, high=max_x, size=K), rd.uniform(low=min_y, high=max_y, size=K)]
    
    #sigmaの初期化
    sigma = np.array([np.eye(D) for k in range(K)])
    #print(sigma)
    
    # 対数尤度の初期化
    pi_Ns, log_likelihood = calc_likelihood(data, mu, sigma, pi, K)
    
    return mu, sigma, pi, pi_Ns, log_likelihood
```

```python
initialize(data, K, D)
```

```python
mu, sigma, pi, pi_Ns, log_likelihood = initialize(data, K, D)
```

## 2. Eステップ

```python
#現在のパラメータを使って、負担率を求める
def E_step(N_s):
    '''Eステップ(負担率を求める)'''
    
    gamma = (N_s.T / np.sum(N_s, axis=1)).T
    
    return gamma
```

```python
E_step(pi_Ns)
```

```python
gamma = E_step(pi_Ns)
```

## 3.Mステップ

```python
def M_step(data, gamma):
    '''負担率を使って、パラメータの値を再計算する'''
    
    #piを計算する
    N_k = np.sum(gamma, axis=0)
    pi = N_k / len(data)
    #print(pi)
    
    #muを計算する
    mu = np.zeros((K,D))
    for k in range(K):
        for i in range(len(data)):
            mu[k] += gamma[i, k] * data[i]
        mu[k] = mu[k] / N_k[k]
    #print(mu)
    
    #sigmaを計算する
    sigma = np.zeros((K,D,D))
    for k in range(K):
        for i in range(len(data)):
            tmp = np.array(data[i] - mu[k])[:, np.newaxis]
            sigma[k] += gamma[i, k]*np.dot(tmp, tmp.T)
        sigma[k] = sigma[k] / N_k[k]     
    #print(sigma)
    
    return mu, sigma, pi
```

```python
M_step(data, gamma)
```

```python
mu_new, sigma_new, pi_new = M_step(data, gamma)
```

## パラメータの評価

```python
def convergence_checking(data, mu_new, sigma_new, pi_new, K, log_likelihood):
    '''収束性を確認するための関数'''
    
    # 対数尤度を計算する
    prev_log_likelihood = log_likelihood
    pi_Ns_new, new_log_likelihood = calc_likelihood(data, mu_new, sigma_new, pi_new, K)
    
    diff = prev_log_likelihood - new_log_likelihood
    
    return pi_Ns_new, new_log_likelihood, diff
```

```python
convergence_checking(data, mu_new, sigma_new, pi_new, K, log_likelihood)
```

推定パラメータによるプロット

## EMアルゴリズ　動作確認

```python
#初期化
mu, sigma, pi, pi_Ns, log_likelihood = initialize(data, K, D)
t = 0

# EM アルゴリズム
while(True):
    t += 1
    
    #Eステップ
    gamma = E_step(pi_Ns)
    
    #Mステップ
    mu_new, sigma_new, pi_new = M_step(data, gamma)
    
    #収束性の確認
    pi_Ns, log_likelihood, diff = convergence_checking(data, mu_new, sigma_new, pi_new, K, log_likelihood)
    
    print("step: %s, diff: %s"%(t, diff))
    if abs(diff) < 0.001:
        break
```

## 可視化と確認

```python
def visualize(data, mu_new, sigma_new, pi_new, gamma, mu_prev, K):
    '''可視化する関数'''
    
    plt.figure(figsize=(12,8))
    for i in range(len(data)):
        plt.scatter(data[i,0], data[i,1], c=gamma[i], alpha=0.5, marker="+")
```
