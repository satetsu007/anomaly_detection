```python
#-*- coding:utf-8 -*-
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import numpy.random as rd
import itertools
```

```python
#データ生成(離散値ベクトル)
rd.seed(0)
T = 500 #データ数(観測数)
x_1 = np.array([np.argmax(i) for i in rd.multinomial(1, [1/6]*6,T)])
x_2 = rd.binomial(1, 0.5, T)

x = np.c_[x_1, x_2]
#x = x_1
print(x)
```

```python
#領域の分割
A = list(itertools.product(set(x_1), set(x_2)))


print(A)

#Given
M = len(A) #セルの数
r = 1 / T #忘却係数
beta = 1  #正の定数
```

```python
#初期化
T_t = np.zeros((T+1,M))
t = 1
prob = np.zeros((T+1, M))
A_det = np.zeros(M)
```

```python
#パラメータの更新
while t <= T:
    #print("x_t:", x[t-1])
    x_t = x[t-1] #
    if isinstance(x_t, np.int64):
        x_t = np.array([x_t])
    
    for i, A_m in enumerate(A):
        delta = 0
        if np.array_equal(x_t, np.array(A_m)):
            delta = 1
            A_det[i] += 1
            
        T_t[t,i] = (1-r) * T_t[t-1,i] + delta
        q = (T_t[t,i] + beta) / ((1 - (1-r)**t)/r + M*beta)
        
        if not A_det[i] == 0:
            prob[t, i] = q / A_det[i]
    
    t += 1
```
